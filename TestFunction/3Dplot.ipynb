```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import lines
import matplotlib.cm as cm


import model_opt
import data_generate
import algo_GD
from ML2_lib import loss
```

```python
f = model_opt.RosenBrock()

x = np.arange(-2, 2, 0.01)
y = np.arange(-1, 2, 0.01)
X, Y = np.meshgrid(x, y)
Z = f.f_opt([X,Y])
```

```python
fdic = {
        "size": 20,
    }

fig = plt.figure(figsize=(8.0, 8.0))

fig.add_subplot(111)
ax = Axes3D(fig)
ax.xaxis._axinfo['label']['space_factor'] = 2.8
ax.set_xlabel("\nw_2",fontdict=fdic)
ax.set_ylabel("\n \n w_1",fontdict=fdic)
ax.tick_params(labelsize = 15,labelright=False)
#ax.set_zlabel("\nf(w)",fontdict=fdic)

ax.plot_surface(Y,X , Z,)
plt.tight_layout
plt.savefig("3dfunction_png/rosenbrock_function.png",bbox_inches="tight")
plt.show()
```

```python


f = model_opt.Bohachevsky()

x = np.arange(-3, 3, 0.001)
y = np.arange(-3, 3, 0.001)
X, Y = np.meshgrid(x, y)
Z = f.f_opt([X,Y])
```
