```python

import numpy as np
import pandas as pd

from ML2_lib import many_trials_SGD, many_trials_SGDAve
from ML2_lib import models
from ML2_lib import plot_set
import matplotlib.pyplot as plt
```

```python
hoge = pd.read_csv("../exp/real_data_exp/real_data_remote_exp/SGD_adult_trial_25_2021-12-16 17:13:04.641742.csv")
```

```python
result = hoge.values
```

```python
hoge.columns
```

```python
fdic = {
    "size": 20,
}

columns = hoge.columns
fig = plt.figure(figsize=(40.0, 8.0))
ax1 = fig.add_subplot(111)


ax1.boxplot(result)
ax1.set_xticklabels(columns, fontsize=15)
ax1.set_title('SGD 25 trials accuracy dataset adult', fontsize=15)
ax1.set_xlabel('learning late - late', fontdict=fdic)
ax1.set_ylabel('accuracy')

plt.savefig("SGD_25_adult_accuracy.png")

plt.show()
```
