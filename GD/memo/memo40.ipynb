```python

import numpy as np
import pandas as pd

from ML2_lib import many_trials_SGD, many_trials_SGDAve
from ML2_lib import models
from ML2_lib import plot_set
import matplotlib.pyplot as plt
import os
import re
```

```python
file_list = os.listdir("../exp/real_data_exp/real_data_remote_exp/CIFAR10/result_/")
file_list
```

```python
fdic = {
    "size": 20,
}
for i in file_list:
    print(i)
    hoge = pd.read_csv(f"../exp/real_data_exp/real_data_remote_exp/CIFAR10/result_/{i}")
    columns = hoge.columns
    fig = plt.figure(figsize=(20.0, 8.0))
    ax1 = fig.add_subplot(111)

    title_tmp = i.split("_")
    title = " ".join(title_tmp[-4:])

    ax1.boxplot(hoge.values)
    ax1.tick_params(axis="x", labelsize=30)
    ax1.tick_params(axis="y", labelsize=30)
    ax1.set_xticklabels(columns, fontsize=20)
    ax1.set_title(title, fontsize=30)
    ax1.set_xlabel('separate num', fontdict=fdic)
    ax1.set_ylabel('accuracy', fontdict=fdic)

    # plt.savefig(f"CIFAR10_result_png//{title}.png")

    plt.show()
```
