```python
import numpy as np
import pandas as pd

from ML2_lib import many_trials_SGD,many_trials_SGDAve
from ML2_lib import models
```

```python
name_path = "data/adult/adult_classification.names"
train_path = "data/adult/adult_classification.csv"
test_path = "data/adult/adult_classification.test"
w_dim = 104
class_num = 2
unit_num = 4
model_1 = models.LinearClassification(w_num=w_dim,c_num=class_num)
model_2 = models.FF_L1(w_num=w_dim,c_num=class_num,unit_num=unit_num)
model_3 = models.FF_L2(w_num=w_dim,c_num=class_num,unit_num=unit_num)
model_list = [model_1,model_2,model_3]
model_name_list = ["Linear", "FF_L1", "FF_L2"]
lr_list = [0.001,0.01,0.1,1]
many_learner = many_trials_SGD.ExManyTrials(train_path=train_path,name_path=name_path,test_path=test_path)
```

```python
result_accuracy, _ = many_learner.ex_many_settings(trial_num=3,model_name_list=model_name_list,lr_list=lr_list,model_list=model_list)
```

```python
pd.DataFrame(np.array(result_accuracy).T,columns=['lr_0.001_model_Linear', 'lr_0.001_model_FF_L1', 'lr_0.001_model_FF_L2', 'lr_0.01_model_Linear', 'lr_0.01_model_FF_L1', 'lr_0.01_model_FF_L2', 'lr_0.1_model_Linear', 'lr_0.1_model_FF_L1', 'lr_0.1_model_FF_L2', 'lr_1_model_Linear', 'lr_1_model_FF_L1', 'lr_1_model_FF_L2']
)
```

```python

many_learner = many_trials_SGDAve.ExManyTrials(train_path=train_path,name_path=name_path,test_path=test_path)
```

```python
result_accuracy, _ = many_learner.ex_many_settings(trial_num=3,model_name_list=model_name_list,lr_list=lr_list,model_list=model_list)
```
