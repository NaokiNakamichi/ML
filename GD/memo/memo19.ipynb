```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from ML2_lib import DC_SGD
from ML2_lib import RV_SGDAve
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn import model_selection
from ML2_lib import models
from ML2_lib import SGDByTorch
```

```python
real_data = load_iris()
x = real_data.data

scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
# x = x.T
```

```python
y = real_data.target
y = y.T

data_train, data_test, label_train, label_test = model_selection.train_test_split(x, y)
w_dim = data_train.shape[1]
class_num = 3
model = models.Net(w_num=w_dim,unit_num=3,c_num=class_num)
# model = models.LinearClassification(4,3)
hoge = SGDByTorch.SGDTorch(lr=1)
```

```python
sgd_model, stack, loss_stack, test_loss_stack = hoge.learn(x=data_train, y=label_train, model=model,class_num=class_num,X_test=data_test,Y_test=label_test)
```

```python
fig = plt.figure(figsize=(10.0, 8.0))
ax1 = fig.add_subplot(111)

ax1.set_title("MSE ")

ax1.plot(loss_stack)
ax1.set_xlabel("step")
plt.show()
```
