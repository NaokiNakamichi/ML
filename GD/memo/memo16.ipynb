```python
from ML2_lib import loss
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from ML2_lib import plot_set


import matplotlib.colors as colors
from matplotlib.cm import ScalarMappable
from mpl_toolkits.mplot3d import Axes3D
```

```python
f = loss.Ackley(d=2, noise_type=None, E_var=1.75, noise_type_f=None, f_E_var=1.75)
```

```python
w1 = np.arange(10)
w2 = np.arange(10)
f.f_opt(w=[w1,w2])
```

```python
w_star = f.w_star
grid_x_max, grid_y_max = 10,10
grid_x_min, grid_y_min = -10,-10
xvals = np.arange(grid_x_min, grid_x_max, 0.01)
yvals = np.arange(grid_y_min, grid_y_max, 0.01)
X, Y = np.meshgrid(xvals, yvals)
Z = f.f_opt([X, Y])

fig, axes = plt.subplots(1, 1, figsize=(6, 6))
# axes.pcolor(X, Y, Z, cmap=plt.cm.rainbow,shading='auto')
ctrx = axes.contour(X, Y, Z,
                    levels=[0, 1,2, 3, 5, 7,8,9,10,15,16,17,18,19,20],  # 等高線の間隔
                    linewidths=1,  # 等高線の幅
                    colors="black"
                    )
axes.clabel(ctrx)
cmap = plt.get_cmap("tab10")
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
Z
```

```python

# カラーバーの設定
fig, axes = plt.subplots(1, 1, figsize=(6, 6))
axes.pcolor(X, Y, Z, cmap=plt.cm.rainbow, shading='auto')
axpos = axes.get_position()
cbar_ax = fig.add_axes([0.9, axpos.y0, 0.03, axpos.height])
norm = colors.Normalize(vmin=Z.min(), vmax=Z.max())
mappable = ScalarMappable(cmap=plt.cm.rainbow, norm=norm)
mappable._A = []
fig.colorbar(mappable, cax=cbar_ax)

# 余白の調整
plt.subplots_adjust(right=0.85)
plt.subplots_adjust(wspace=0.1)
plt.show()
```
