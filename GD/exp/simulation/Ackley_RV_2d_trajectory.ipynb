```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from ML2_lib import loss
from ML2_lib import plot_set
from ML2_lib import RV_SGDAve
```

```python
d = 2
trial_num = 100
lr = 0.1
noise = "lognormal"
c = 1
E_var = 0.7
n = 10000
k_list = [1, 2, 4, 5, 10, 20]
son = loss.Ackley(d=d, noise_type=noise, E_var=E_var)
```

```python
w_init = np.array([5, 5])

title = f"RVSGD_Ackley_trajectory_noise_{noise}_D_{d}_sample_num{n}_noise_va_{E_var}_w_init_{w_init[0]}_{w_init[1]}"
RV = RV_SGDAve.RVSGDByW(model_opt=son, c=1, n=n, lr=lr)
k_core_list, k_selected_index, k_w_rv = RV.k_all_transition(k_list=k_list, w_init=w_init)
plot_set.multiple_w_value_2d_k_candidates_contour(k_list=k_list, k_list_core_store=k_core_list, _t_max=n,
                                                  k_selected_index=k_selected_index, f=son, levels=[2,5,10,15,17,20],title=title,saving_png=False)
```

```python
RV = RV_SGDAve.RVSGDByW(model_opt=son, c=1, n=n, lr=lr)
w_init = np.array([30, 30])
k_core_list, k_selected_index, _ = RV.k_all_transition(k_list=k_list, w_init=w_init)
plot_set.multiple_w_value_2d_k_candidates_contour(k_list=k_list, k_list_core_store=k_core_list, _t_max=n,
                                                  k_selected_index=k_selected_index, f=son, levels=[2,5,10,15,17,20])
```
